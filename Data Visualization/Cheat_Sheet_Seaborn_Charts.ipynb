#import

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
olympic = pd.read_csv('/content/120-years-of-olympic-history-athletes-and-results/athlete_events.csv')
healthcare = pd.read_csv('/content/av-healthcare-analytics-ii/healthcare/train_data.csv')
covid_india = pd.read_csv('/content/covid19-in-india/covid_19_india.csv')
google_playstore = pd.read_csv('/content/google-play-store-apps/googleplaystore.csv')
iris = pd.read_csv('/content/iris/Iris.csv')
corona_virus = pd.read_csv('/content/novel-corona-virus-2019-dataset/covid_19_data.csv')
youtube = pd.read_csv('/content/youtube-new/USvideos.csv')
shootings = pd.read_csv('/content/us-police-shootings/shootings.csv')
```

# Basic Sctter Plot

```python
sns.scatterplot(x = 'SepalLengthCm', y = 'SepalWidthCm', data=iris)
plt.title('Sepal Length vs. Sepal Width')
```

Scatter Plot - category

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=iris)
plt.title('Sepal Length vs. Sepal Width')
```

Scatter Plot with Regression Fit

```python
sns.regplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title('Sepal Length vs. Sepal Width with Regression Fit')
```

Control shape of elements in Scatter Plot

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue ='Species', data=iris, marker='+')
plt.title('Sepal Length vs. Sepal Width')
```

Using color Palletes in Scatter Plot

```python
sns.scatterplot(x='Age', y='Height', hue = 'Sex', data=olympic, palette='Blues')
plt.title('Age vs. Height')
```

```python
sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='PuOr')
plt.title('Age vs. Height')
```

```python
sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette = 'Set1')
plt.title('Age vs. Height')
```

#Line Plot

Basic Line Plot

```python
sns.lineplot(x='Age', y='Height', data=olympic)
```

Line Plot - Category

```python
sns.lineplot(x='Age', y='Height', data=olympic, hue='Sex')
```

```python
sns.lineplot(x='Age', y='Height', data=olympic, estimator=None)
```

# Histogram Plot

Basic Histogram Plot

```python
f, ax = plt.subplots(figsize = (7,5))
sns.despine(f)
sns.distplot(olympic['Age'])
```

Histogram Plot : Control Number of Bins

```python
f, ax = plt.subplots(figsize = (7,5))
sns.despine(f)
sns.distplot(olympic['Age'], bins = 20)
```

Histogram Plot : control Density

```python
f, ax = plt.subplots(figsize = (7,5))
sns.despine(f)
sns.distplot(olympic['Age'], kde=False)
```

# 1D Density Plot

Basic 1D Density Plot

```python
f, ax = plt.subplots(figsize=(7,5))
sns.despine(f)
sns.kdeplot(iris['PetalLengthCm'])
```

Density Plot with Shade

```python
f, ax = plt.subplots(figsize = (7,5))
sns.despine(f)
sns.kdeplot(iris['PetalLengthCm'], shade=True)
```

Horizontal Density Plot

```python
f, ax = plt.subplots(figsize = (7,5))
sns.despine(f)
sns.kdeplot(iris['PetalLengthCm'], vertical=True)
```

Horizontal Density Plot with Shade

```python
f, ax = plt.subplots(figsize=(7,5))
sns.despine(f)
sns.kdeplot(iris['PetalWidthCm'], vertical = True, shade=True)
```

Multi Variable Density Plot

```python
f, ax = plt.subplots(figsize = (7,5))
sns.despine(f)
sns.kdeplot(iris['PetalWidthCm'], shade=True)
sns.kdeplot(iris['PetalLengthCm'], shade=True)
plt.show()
```

Multi Variable Density Plot : Customize Color

```python
f, ax = plt.subplots(figsize=(7,5))
sns.despine(f)
sns.kdeplot(iris['SepalLengthCm'], shade = True, color = 'r')
sns.kdeplot(iris['SepalWidthCm'], shade=True, color='m')
plt.show()
```

# 2D Density Plot

Basic 2D Density Plot

```python
sns.kdeplot(x = iris['SepalLengthCm'], y=iris['SepalWidthCm'])
plt.title('Sepal Length vs Sepal Width 2D Density Plot')
```

```python
sns.kdeplot(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'], cmap='Reds', shade=True, bw=.15)
```

```python
sns.kdeplot(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'], cmap='Blues', shade=True, shade_lowest=True,)
```

# Bar Plot

Basic Bar Plot

```python
f, ax = plt.subplots(figsize=(7,5))
sns.despine(f)
sns.barplot(x='Species', y='SepalLengthCm', data=iris, palette='magma')
```

```python
f, ax = plt.subplots(figsize= (7,5))
sns.despine(f)
sns.barplot(x='SepalLengthCm', y='Species', data=iris, palette='magma', orient='h')
```

# Seaborn Themes

Seaborn Dark Grid Theme

```python
sns.set_style('darkgrid')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title('Seaborn Dark Grid Theme')
```

Seaborn Dark Theme

```python
sns.set_style("dark")
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title("Seaborn Dark Theme")
```

Seaborn White Grid Theme

```python
sns.set_style('whitegrid')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title('Seaborn White Grid Theme')
```

Seaborn White Theme

```python
sns.set_style('white')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title('Seaborn White Theme')
```

Seaborn Ticks Theme

```python
sns.set_style('ticks')
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', data=iris)
plt.title('Seaborn Ticks Theme')
```
